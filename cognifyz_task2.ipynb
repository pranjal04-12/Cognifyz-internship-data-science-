```python
import pandas as pd

df = pd.read_csv('/content/Dataset  (2).csv')

print(df.head())
```

```python
print("Shape of dataset:", df.shape)
```

```python
print(df.isnull().sum())
```

```python
df['Cuisines'].fillna("Unknown", inplace=True)
```

```python
print(df.dtypes)
```

```python
df['Has Table booking'] = df['Has Table booking'].map({'Yes':1, 'No':0})
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes':1, 'No':0})
```

```python
print(df['Aggregate rating'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Aggregate rating'].hist()
plt.title("Aggregate Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
print(df.describe())
```

```python
print(df['Country Code'].value_counts())
```

```python
print(df['City'].value_counts())
```

```python
print(df['Cuisines'].value_counts())
```

```python
top_cities = df['City'].value_counts().head(10)
print(top_cities)
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)
print(top_cuisines)
```

```python
import matplotlib.pyplot as plt

top_cities.plot(kind='bar')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cuisines.plot(kind='bar')
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
