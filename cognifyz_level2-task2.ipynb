```python
import pandas as pd
df = pd.read_csv('/content/Dataset  (2).csv')

import folium

# Map center (average location)
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]

# Create map
m = folium.Map(location=map_center, zoom_start=5)

# Add some points (limit for performance)
for i in range(100):
    folium.Marker(
        location=[df.iloc[i]['Latitude'], df.iloc[i]['Longitude']],
        popup=df.iloc[i]['Restaurant Name']
    ).add_to(m)

m
```

# **cognifyz_level2_task1**
# **Task: Table Booking and Online Delivery**

Determine the percentage of restaurants that
offer table booking and online delivery.

Compare the average ratings of restaurants
with table booking and those without.

Analyze the availability of online delivery
among restaurants with different price ranges.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Dataset  (2).csv')

# Handle missing values in 'Cuisines' (as done in previous steps)
df['Cuisines'].fillna("Unknown", inplace=True)

# Convert 'Yes'/'No' to 1/0 for 'Has Table booking' and 'Has Online delivery' (as done in previous steps)
df['Has Table booking'] = df['Has Table booking'].map({'Yes':1, 'No':0})
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes':1, 'No':0})

# Table booking percentage
table_booking_percent = df['Has Table booking'].mean() * 100

# Online delivery percentage
online_delivery_percent = df['Has Online delivery'].mean() * 100

print("Table Booking %:", table_booking_percent)
print("Online Delivery %:", online_delivery_percent)
```

```python
avg_rating_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()

print(avg_rating_booking)
```

```python
delivery_price = df.groupby('Price range')['Has Online delivery'].mean()

print(delivery_price)
```

```python
import matplotlib.pyplot as plt

df['Has Table booking'].value_counts().plot(kind='bar')
plt.title("Table Booking Availability")
plt.show()
```

```python
delivery_price.plot(kind='bar')
plt.title("Online Delivery vs Price Range")
plt.xlabel("Price Range")
plt.ylabel("Availability")
plt.show()
```
